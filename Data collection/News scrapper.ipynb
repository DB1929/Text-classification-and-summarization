```python
import pandas as pd
import newspaper
from newspaper import Article
import os
```

```python
df = pd.read_csv('news-aggregator.csv')

df.head()
```

```python
df1 = df.loc[df['CATEGORY'] == 'health']
df2 = df.loc[df['CATEGORY'] == 'entertainment']
df3 = df.loc[df['CATEGORY'] == 'business']
df4 = df.loc[df['CATEGORY'] == 'technology']
```

```python
# change the values df1.iloc[1000:2000,:]
df1 = df1.iloc[:1000,:]
df2 = df2.iloc[:1000,:]
df3 = df3.iloc[:1000,:]
df4 = df4.iloc[:1000,:]
```

```python
def download_content_summary(URL):
    try:
        print("Downloading from..",URL)
        article = Article(URL)
        article.download()
        article.parse()
        if(len(article.text) > 150 and len(article.text)<6000):
            article.nlp()
            return [article.text, article.summary]
        
    except Exception:
        pass
```

```python
df1['content and summary'] = df1['URL'].apply(download_content_summary)
df1.to_csv('health_csv',encoding="utf-8-sig",index = False)
```

```python
df2['content and summary'] = df2['URL'].apply(download_content_summary)
df2.to_csv('entertainment.csv',encoding="utf-8-sig",index = False)
```

```python
df3['content and summary'] = df3['URL'].apply(download_content_summary)
df3.to_csv('business.csv',encoding="utf-8-sig",index = False)
```

```python
df4['content and summary'] = df4['URL'].apply(download_content_summary)
df4.to_csv('technology.csv',encoding="utf-8-sig",index = False)
```
